# Importing Necessary Libararies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```


# Load Dataset

```python
df= pd.read_csv('D:/emails.csv')
df.head()
```

# Handle Missing Values

```python
df = df.dropna()
```

```python
df.head()
```

# Extracting Features

```python
X=df.iloc[:,1:-1]
Y=df.iloc[:,-1]
```

# Split Data into Train & Test Dataset

```python
X_train, X_test, y_train, y_test= train_test_split(X,Y, test_size=0.2 , random_state=42)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

# Models Training

```python
#creating objects of every models
nb = MultinomialNB()
rf= RandomForestClassifier()
gb = GradientBoostingClassifier()
lr = LogisticRegression()
```

# NAIVE BAYES MODEL TRAIN

```python
nb.fit(X_train,y_train)
```

# Check Accuracy of Naive Bayes Model

```python
y_pred = nb.predict(X_test)
accuracy_score(y_test, y_pred)
```

# RandomForest Model Train & Accuracy Check 

```python
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
accuracy_score(y_test, y_pred)
```

# GradientBoosting Model Train & Accuracy Check

```python
gb.fit(X_train,y_train)
```

```python
y_pred = gb.predict(X_test)
accuracy_score(y_test, y_pred)
```

# LogisticRegression Model Train & Accuracy Check

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
accuracy_score(y_test, y_pred)
```

# Hyperparameter Tunning NaiveBayes & LogisticRegression Models

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
alpha = [0.1,0.5,1.0,1.5,2.0]
fit_prior=[True,False]
```

```python
param_grid = {
    'alpha': alpha,
    'fit_prior': fit_prior
}
```

```python
nb_grid = GridSearchCV(estimator=nb, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
nb_grid.fit(X_train,y_train)
```

```python
nb_grid.best_params_
```

```python
nb_grid.best_score_
```

# RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform
```

```python
param_distributions = {
    'alpha': uniform(0.1, 2.0), 
    'fit_prior': [True, False]
}
```

```python
random_nb_grid = RandomizedSearchCV(estimator=nb, param_distributions=param_distributions, n_iter=100, cv=5, scoring='accuracy', random_state=42)
```

```python
random_nb_grid.fit(X_train, y_train)
```

```python
random_nb_grid.best_params_
```

```python
random_nb_grid.best_score_
```

# LogisticRegression GridSearchCV

```python
# Hyperparameter tuning for LogisticRegression with GridSearchCV
# Regularization strength
C=[0.01, 0.1, 1, 10, 100] 
# Solver
solver= ['liblinear', 'lbfgs'] 
# Penalty
penalty= ['l1', 'l2']  
# Maximum iterations
max_iter=[100, 200, 300] 
```

```python
 param_grid_lr = {
    'C': C,  
    'solver': solver, 
    'penalty': penalty, 
    'max_iter': max_iter  
}
```

```python
lr_grid = GridSearchCV(estimator=lr, param_grid=param_grid_lr, cv=5, scoring='accuracy')
```

```python
lr_grid.fit(X_train, y_train)
```

```python
lr_grid.best_params_
```

```python
 lr_grid.best_score_
```

# LogisticRegression RandomizedSearchCV

```python
from scipy.stats import uniform, loguniform
```

```python
param_distributions_lr = {
    'C': loguniform(0.01, 100),  # Regularization strength, log-uniform distribution
    'solver': ['liblinear', 'lbfgs'],  # Solver
    'penalty': ['l1', 'l2'],  # Penalty
    'max_iter': [100, 200, 300]  # Maximum iterations
}
```
